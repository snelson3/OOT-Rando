```python
from oot import Oot

ACTOR_OVERLAY_TABLE = int('0x00B5E490', 0) 
AOT_SIZE = 470 * 32 # Each record is 32 bytes long
```

```python
oot = Oot()
```

```python
import pandas
```

```python
# Get the ActorOverlayTable
table_list = []
toHex = lambda l: [hex(h)[2:] for h in l]
for b in range(ACTOR_OVERLAY_TABLE, ACTOR_OVERLAY_TABLE+AOT_SIZE, 32):
    record = {
        "record_location": b,
        "start_rom_address": oot.data[b:b+4],
        "end_rom_address": oot.data[b+4:b+8],
        "start_ram_address": oot.data[b+8:b+12],
        "end_ram_address": oot.data[b+12:b+16],
        "ram_address_of_overlay": oot.data[b+16:b+20],
        "init_file": oot.data[b+20:b+24],
        "filename": oot.data[b+24:b+28],
        "allocation_type": oot.data[b+28:b+30],
        "num_loaded": oot.data[b+32]
    }
    table_list.append(record)
pandas.DataFrame([[hex(r["record_location"])[2:],
                   toHex(r["start_rom_address"]),
                   toHex(r["end_rom_address"]),
                   toHex(r["init_file"])] for r in table_list],
                columns=["record_loc", "start_rom", "end_rom", "init"]).head(328)


```
